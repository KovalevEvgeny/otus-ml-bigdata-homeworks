```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()

%matplotlib inline
```

```python
df = pd.read_csv('data/Dataset.csv')
df.head()
```

```python
plt.figure(figsize=(11, 8))
sns.histplot(data=df, x=np.log1p(df['output']), bins=20, kde=True)
plt.title('Histogram of log(output) values')
plt.xlabel('log(output)')
plt.ylabel('counts')
plt.savefig('images/histogram_log-output.jpg', dpi=200)
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.savefig('images/correlations.jpg', dpi=200)
```
